```python
import requests
from bs4 import BeautifulSoup, Tag
```

```python
url = "https://ycharts.com/indicators/us_average_retail_price_of_electricity_total"

headers = {
    "User-Agent": "Firefox"
}

res = requests.get(url, headers=headers)
res
```

```python
html = res.text
soup = BeautifulSoup(html)
tables = soup.find_all("table")
tables
```

```python
table: Tag
for table in tables:
    for tag in table.children:
        print(tag.text)
```
